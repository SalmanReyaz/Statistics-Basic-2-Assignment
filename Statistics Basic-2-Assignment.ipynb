# `Q1. What are the three measures of central tendency?`

The Three Measures of Central Tendency are:---

# 1.`MEAN OR AVERAGE:- The mean, also known as the average, is calculated by summing up all the values in a dataset and then dividing the sum by the total number of values. It represents the "center" of the data in terms of the arithmetic average. The formula for calculating the mean is:`

# `Mean = (Sum of all values) / (Number of values)`

`For example, if you have the dataset {2, 3, 5, 7, 10}, the mean is (2 + 3 + 5 + 7 + 10) / 5 = 5.4.`

# 2. MEDIAN:- The median is the middle value of a dataset when the values are arranged in ascending or descending order. If there is an even number of values, the median is the average of the two middle values. It's a measure of central tendency that is not affected by extreme values or outliers in the dataset. To find the median:

# Arrange the values in ascending or descending order.
`If the number of values is odd, the median is the middle value.
If the number of values is even, the median is the average of the two middle values.
For example, for the dataset {2, 3, 5, 7, 10}, when arranged in ascending order, the median is 5`.

    
# 3. MODE:-The mode is the value or values that appear most frequently in a dataset. 

`For example, in the dataset {2, 3, 5, 7, 5, 10, 2, 7}, the mode is both 2 and 5, as they each appear twice, which is more frequently than any other value.`

# Positive Skewness (Right Skewness):

# In a positively skewed distribution, the tail on the right-hand side (the higher values) is longer or fatter than the left-hand side tail. 

# The majority of the data points are concentrated on the left side, and there are relatively few extreme values on the right side.





`Negative Skewness (Left Skewness):`

`In a negatively skewed distribution, the tail on the left-hand side (the lower values) is longer or fatter than the right-hand side tail`


`The majority of the data points are concentrated on the right side, and there are relatively few extreme values on the left side`.



__Symmetric (No Skewness):__

## `In a symmetric distribution, there is an equal balance of data on both sides of the mean, resulting in a bell-shaped or approximately normal distribution.`

## `The mean and median are approximately equal in a symmetric distribution.`

# Q10. Explain the difference between covariance and correlation. How are these measures used in statistical analysis?

- Covariance measures the degree to which two variables change together. It quantifies whether an increase (or decrease) in one variable corresponds to an increase (or decrease) in the other variable

`The value of covariance lies in the range of -∞ and +∞.`


`Covariance values are not standardized.






Correlation is a standardized measure of the strength and direction of the linear relationship between two variables. It quantifies how closely the data points cluster around a linear regression line

`Correlation is limited to values between the range -1 and +1`


`Correlation values are standardized`

# Covariance in Statistical Analysis:

# Identifying Relationships: Covariance is used to identify whether two variables are positively or negatively related. If the covariance is positive, it suggests that the variables tend to increase or decrease together, while a negative covariance indicates they move in opposite directions.


# Correlationin Statistical Analysis:

### Hypothesis Testing: Correlation is used for hypothesis testing to determine whether the observed correlation coefficient is statistically significant. This helps researchers assess whether the observed relationship is likely due to chance or if it represents a real association.

### Model Building: Correlation is often used in regression analysis and model building. It helps in selecting relevant predictor variables that have a strong linear relationship with the target variable, improving the predictive accuracy of the model.


### Portfolio Management: In finance, correlation is used to assess the degree to which assets within a portfolio are correlated. Understanding correlations between assets helps in diversification strategies to manage risk effectively.

# Q11. What is the formula for calculating the sample mean? Provide an example calculation for a dataset?

```python
## lets say example of sample mean:--

data = [2,3,4,5,35,66,44,6,4,6,3,55,66,33,4,5,55,6,44,5]

# Calculate the sample mean
sample_mean = sum(data) / len(data)

# Print the result
print("Sample Mean:", sample_mean)
```

```python
def calculate_sample_mean(data):
    # Calculate the sample mean
    sample_mean = sum(data) / len(data)
    return sample_mean

# Define a dataset
data = [1,2,3,4,5,6]

# Call the function to calculate the sample mean
result = calculate_sample_mean(data)

# Print the result
print("Sample Mean:", result)
```

# Q12. For a normal distribution data what is the relationship between its measure of central tendency?

### In a normal distribution, which is also known as a Gaussian distribution or bell curve, there is a specific relationship between its measures of central tendency, which include the mean, median, and mode:


`Mean (μ): In a perfectly symmetrical normal distribution, the mean is located at the center of the distribution. It is equal to the median and the mode. This is one of the defining characteristics of a normal distribution.`


__Median (also μ): The median of a normal distribution is the same as its mean (μ). Since a normal distribution is symmetric, with the data values roughly equally distributed on both sides of the mean, the median is also at the center.__

## Mode (also μ): The mode of a perfectly symmetrical normal distribution is also the same as its mean (μ). In a normal distribution, the peak of the distribution curve (the mode) occurs exactly at the mean.

# Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

### Outliers can have a significant impact on measures of central tendency (mean, median, and mode) and measures of dispersion (variance and standard deviation). Outliers are data points that are significantly different from the majority of the data in a dataset and can distort the summary statistics. Here's how outliers affect these measures, along with an example:`




# `Measures of Central Tendency:`

# Mean: Outliers can heavily influence the mean, pulling it in the direction of the outlier. If you have a dataset with a very high or very low outlier, the mean can be skewed in that direction.

# Median: The median is less affected by outliers. It is resistant to extreme values because it represents the middle value when the data is sorted, rather than relying on the actual values themselves.

# Mode: The mode represents the most frequently occurring value in the dataset. Outliers typically do not affect the mode unless they are repeated multiple times, which can result in bimodal or multimodal distributions.


#`Measures of Dispersion:`

## Variance: Outliers can significantly increase the variance because the squared differences between data points and the mean are sensitive to extreme values.

## Standard Deviation: Like variance, the standard deviation is influenced by outliers. It measures the average distance of data points from the mean, so extreme values can lead to larger standard deviations.





```python
import numpy as np
import statistics as stats

# Create a dataset with outliers
data_with_outliers = [10,12,14,12,15,23,4,780,800]

# Calculate mean, median,mode, variance and standard deviation
mean_with_outliers = np.mean(data_with_outliers)
median_with_outliers = np.median(data_with_outliers)
mode_with_outliers   = stats.mode(data_with_outliers)
variance_with_outliers = np.var(data_with_outliers)
std_dev_with_outliers = np.std(data_with_outliers)



print("Mean:", mean_with_outliers)
print("Median:", median_with_outliers)
print("Mode:",mode_with_outliers)
print("Variance:", variance_with_outliers)
print("Standard Deviation:", std_dev_with_outliers)
```
